```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from matplotlib.colors import hsv_to_rgb
import os
```

```python
# define x and y vectors

xs = np.arange(0., 1., 0.01)
ys = np.arange(0., 1., 0.01)

# stability function

def stability(x,y):
    return 1 - x

def plasticiy(x,y):
    return np.sqrt(x**2 + y**2)

# create meshgrid
XX, YY = np.meshgrid(xs, ys)
stabilities = stability(XX,YY)
plasticities = plasticiy(XX,YY)

# norm both
stabilities = stabilities / np.max(stabilities)
plasticities = plasticities / np.max(plasticities)
```

```python
if not os.path.exists('plots/conceptual'):
    os.makedirs('plots/conceptual')

plt.figure(figsize=(4,4))
plt.imshow(stabilities, cmap='Blues', origin='lower', alpha=1)
plt.contour(stabilities, cmap='binary', origin='lower')
sns.despine()
plt.xticks([10, 90], [0, 1])
plt.yticks([10, 90], [0, 1])
plt.xlabel(r'$\Delta h CC^{\top}$')
plt.ylabel(r'$\Delta h NN^{\top}$')
plt.title('Stability')
plt.savefig('plots/conceptual/stability.png', dpi=300, bbox_inches='tight')


plt.figure(figsize=(4,4))
plt.imshow(plasticities, cmap='Oranges', origin='lower')
plt.contour(plasticities, cmap='binary', origin='lower')
sns.despine()
plt.xticks([10, 90], [0, 1])
plt.yticks([10, 90], [0, 1])
plt.xlabel(r'$\Delta h CC^{\top}$')
plt.ylabel(r'$\Delta h NN^{\top}$')
plt.title('Plasticity')
plt.savefig('plots/conceptual/plasticity.png', dpi=300, bbox_inches='tight')


plt.figure(figsize=(4,4))
plt.imshow(stabilities + plasticities, cmap='Purples', origin='lower')
plt.contour(stabilities + plasticities, cmap='binary', origin='lower')
sns.despine()
plt.xticks([10, 90], [0, 1])
plt.yticks([10, 90], [0, 1])
plt.xlabel(r'$\Delta h CC^{\top}$')
plt.ylabel(r'$\Delta h NN^{\top}$')
plt.title('Plasticity + Stability')
plt.savefig('plots/conceptual/plasticity_stability.png', dpi=300, bbox_inches='tight')






# %%
```
